# Duplicate Image Finder - Android APK Builder
在 Google Colab 上一键构建 APK，完全免费，无需安装任何东西。

## 使用步骤
1. 点击菜单栏: 运行时 → 全部运行
2. 等待 15-30 分钟（首次需要下载 SDK）
3. 构建完成后 APK 自动下载到本地

```
# 步骤1: 安装依赖
!sudo apt update -qq
!sudo apt install -y -qq python3-pip openjdk-17-jdk git zip unzip autoconf libtool libssl-dev libffi-dev zlib1g-dev 2>&1 | tail -3
!pip install buildozer cython 2>&1 | tail -3
print('环境准备完毕!')
```

```
# 步骤2: 上传项目文件
import os

# 克隆或创建项目
PROJECT_DIR = '/content/DuplicateImageFinder'
os.makedirs(PROJECT_DIR, exist_ok=True)
os.makedirs(f'{PROJECT_DIR}/src/ui', exist_ok=True)
os.makedirs(f'{PROJECT_DIR}/assets', exist_ok=True)
print(f'项目目录: {PROJECT_DIR}')
print('请手动上传以下文件 (左侧文件浏览器):')
print('  main.py -> /content/DuplicateImageFinder/')
print('  buildozer.spec -> /content/DuplicateImageFinder/')
print('  src/*.py -> /content/DuplicateImageFinder/src/')
print('  src/ui/*.py -> /content/DuplicateImageFinder/src/ui/')
```

```
# 步骤3: 构建 APK
%cd /content/DuplicateImageFinder
!buildozer android debug 2>&1
print('构建完成!')
```

```
# 步骤4: 下载 APK
from google.colab import files
import glob

apk_files = glob.glob('/content/DuplicateImageFinder/bin/*.apk')
if apk_files:
    for apk in apk_files:
        print(f'下载: {apk}')
        files.download(apk)
else:
    print('未找到APK文件，请检查构建日志')
```
